```python
import cv2
from darkflow.net.build import TFNet
import numpy as np
import time
from time import time as timer

import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg'
```

```python
option = {
    'model': 'cfg/yolo.cfg',
    'load': 'bin/yolo.weights',
    'threshold': 0.15,
    'gpu': 0.85
}

tfnet = TFNet(option)
```

```python
capture = cv2.VideoCapture('/media/cfchen/956df7bc-562e-4f24-8339-fd0b67f98888/Downloaded/VideosHPB/HPB_1.mp4')
#capture = cv2.VideoCapture('../VideosHPB/IMAG0011.mp4')
colors = [tuple(255 * np.random.rand(3)) for i in range(5)]  #range(100)
```

```python
for color in colors:
    print(color)
```

```python
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, 30.0, (1920,1080))
```

## Pre-process (a)

```python
count = 0
while(True):
    stime = time.time()
    ret, frame = capture.read()

    if ret == False:
        break;
    else:
        count = count + 1

    # B & W image (using red channel only)
    #frame = np.stack((frame[:,:,0],frame[:,:,0],frame[:,:,0]),axis=2)
    results = tfnet.return_predict(frame)
    #print(results)

    if ret:
        for color, result in zip(colors, results):
            tl = (result['topleft']['x'], result['topleft']['y'])
            br = (result['bottomright']['x'], result['bottomright']['y'])
            brX = result['bottomright']['x']
            brY = result['bottomright']['y']
            label = result['label']
            confidence = result['confidence']
            confidence = round(confidence, 1)
            frame = cv2.rectangle(frame, tl, br, color, 7)
            #frame = cv2.putText(frame, label, tl, cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)
            #frame = cv2.putText(frame, label, (brX - 50, brY + 20), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
            frame = cv2.putText(frame, label + ' ' + str(confidence), tl, cv2.FONT_HERSHEY_COMPLEX, 0.65, (0, 255, 0), 2)

        out.write(frame)
        cv2.imshow('frame', frame)
        #
        print("Shape of Frame: {}".format(frame.shape))
        print("Type of Frame:  {}".format(type(frame)))
        print("Results:        {}".format(result))
        print('FPS         {:.1f}'.format(1 / (time.time() - stime)))  # Frame per seconds
        #
        if (cv2.waitKey(1) & 0xFF == ord('q')) or (ret == False):
            break

print('++++++++ Video End ++++++++')
  
capture.release()
out.release()
cv2.destroyAllWindows()
```

## Pre-process (b)

```python
count = 0
while(True):
    stime = time.time()
    start = timer()
    ret, frame = capture.read()

    if ret == False:
        break;
    else:
        count = count + 1

    frame = np.stack((frame[:,:,0],frame[:,:,0],frame[:,:,0]),axis=2)
    #results = tfnet.return_predict(frame)
    #
    preprocessed = tfnet.framework.preprocess(frame)
    feed_dict = {tfnet.inp: [preprocessed]}
    net_out = tfnet.sess.run(tfnet.out,feed_dict)[0]

    processed = tfnet.framework.postprocess(net_out, frame, False)
    #
    out.write(processed)
    cv2.imshow('frame', processed)
    #
    print("Shape of Frame: {}".format(frame.shape))
    print("Type of Frame:  {}".format(type(frame)))
    #print("Results:        {}".format(processed))
    print('FPS {:.1f}'.format(1 / (time.time() - stime)))  # Frame per seconds
    print('Elapsed time = ' + str(timer() - start) + ' s')
    #
    if (cv2.waitKey(1) & 0xFF == ord('q')) or (ret == False):
        break

print('++++++++ Video End ++++++++')

capture.release()
out.release()
cv2.destroyAllWindows()
```

## Pre-process (c)

```python
count = 0
while(True):
    stime = time.time()
    start = timer()
    ret, frame = capture.read()

    if ret == False:
        break;
    else:
        count = count + 1

    #frame = np.stack((frame[:,:,0],frame[:,:,0],frame[:,:,0]),axis=2)
    #results = tfnet.return_predict(frame)
    #
    h, w, _ = frame.shape
    
    preprocessed = tfnet.framework.preprocess(frame)
    feed_dict = {tfnet.inp: [preprocessed]}
    net_out = tfnet.sess.run(tfnet.out, feed_dict)[0]

    processed = tfnet.framework.postprocess(net_out, frame, False)
    
    ###############
    # Boxes Info  #
    ###############
    outc2 = tfnet.sess.run(tfnet.out, feed_dict)[0]
    boxes = tfnet.framework.findboxes(outc2)

    threshold = tfnet.FLAGS.threshold

    boxesInfo = list()
    #
    for box in boxes:
        tmpBox = tfnet.framework.process_box(box, h, w, threshold)
        if tmpBox is None:
            continue
        boxesInfo.append({
            "label": tmpBox[4],
            "confidence": tmpBox[6],
            "topleft": {
                "x": tmpBox[0],
                "y": tmpBox[2]},
            "bottomright": {
                "x": tmpBox[1],
                "y": tmpBox[3]}
        })
    #
    print(boxesInfo)
    #
    out.write(processed)
    cv2.imshow('frame', processed)
    #
    print("Shape of Frame: {}".format(frame.shape))
    print("Type of Frame:  {}".format(type(frame)))
    #print("Results:        {}".format(processed))
    print('FPS {:.1f}'.format(1 / (time.time() - stime)))  # Frame per seconds
    print('Elapsed time = ' + str(timer() - start) + ' s')    
    if (cv2.waitKey(1) & 0xFF == ord('q')) or (ret == False):
        break

print('++++++++ Video End ++++++++')

capture.release()
out.release()
cv2.destroyAllWindows()
```
