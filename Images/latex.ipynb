# My favorite equations

## Schroedinger Eq.

$$H\,\psi({\bf{x}}, t)\,=\,i\,\hbar\,\frac{\partial}{\partial t}\,  \psi({\bf{x}}, t)$$

## Dirac Eq.

$$i\,\gamma^\mu \partial_\mu\, \psi - m\, \psi \,=\,0$$

## QCD DGLAP or AP equations

$$\frac{d}{dt} q_i(x,t) \,=\,\frac{\alpha_s(Q)}{2\,\pi}[q_i\otimes P_{qq}] + \frac{\alpha_s(Q)}{2\,\pi}[g\otimes P_{qg}]$$

$$[q \otimes P] \,=\,[P \otimes q]\,=\, \int_x^1 dy\,\frac{q(y, t)}{y} \cdot P\left(\frac{x}{y}\right)$$

$$\frac{d}{dt} g(x,t) \,=\,\frac{\alpha_s(Q)}{2\,\pi}\left[ \sum_i\left(q_i + {\bar q}_i\right) \otimes P_{gq}\right] + \frac{\alpha_s(Q)}{2\,\pi}[g\otimes P_{gg}]$$

# Simple plotting example

This is a plot of the $sinc$ function

$$f(x)\,=\,\frac{\sin(x)}{x}$$

```python
%matplotlib inline
%config InlineBackend.figure_format ='svg'

import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)
plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size':13})

plt.yticks([0, 1],['$0$', '$1$'])
plt.xticks([-4*pi, -2*pi, 0, 2*pi, 4*pi],
          ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'])
#plt.axhline(0.5, color='red', lw=0.5)
plt.axhline(0, color='green', lw=0.5)
plt.plot(x, np.sin(x)/x, label=r'$f(x)\,=\,\frac{\sin(x)}{x}$')
plt.ylim(-0.5, 1.1)
plt.title('The $sinc$ function')
plt.legend(loc='best', fontsize=15)
plt.show()
```
