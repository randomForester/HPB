# My favorite equations

### Schroedinger Eq.

$$H\,\psi({\bf{x}}, t)\,=\,i\,\hbar\,\frac{\partial}{\partial t}\,  \psi({\bf{x}}, t)$$

### Dirac Eq.

$$i\,\gamma^\mu \partial_\mu\, \psi - m\, \psi \,=\,0$$

### The differential parton luminosities (1)

$$\frac{d{\cal L}^{gg}}     {d\tau}   = g\otimes     g(\tau,Q^2)$$

### The differential parton luminosities (2)

$$\frac{d{\cal L}^{gq}}     {d\tau}   = g\otimes     q(\tau,Q^2) +      q\otimes g(\tau,Q^2)$$

### The differential parton luminosities (3)

$$\frac{d{\cal L}^{q\bar q}}{d\tau}   = q\otimes\bar q(\tau,Q^2) + \bar q\otimes q(\tau,Q^2)$$

### The notation $\otimes$ is given by

$$f\otimes g \left(x,\mu^2\right) \,=\, \int_x^1 \frac{dz}{z} \,\,f\left(z,\mu^2\right) \, g\left(\frac{x}{z},\mu^2\right)$$

where $q \,(\bar q)$ and $g$ are PDFs of quark (antiquark) and gluon.

The PDFs belong to the long-distance physics process of the scattering and hence belong to the non-perturbative part. But their evolutions follow the master equation of QCD: DGLAP equation, by which we can predict how PDFs evolve from one scale to another scale.

### QCD DGLAP equation

$$t\frac{d}{dt}q_a(x,t)\,=\,\frac{\alpha_s(t)}{2\pi}\int_x^1 \frac{d\xi}{\xi}
P_{ab}\left(\frac{x}{\xi},\alpha_s(t) \right) q_b(\xi,t)$$

where $t=\mu^2$ and $P_{ab}$ is the splitting function. More generally, the DGLAP equation in $(2\,n_f + 1 )$-dimension in the flavour space (flavour $=$ quarks, antiquarks and gluons) is:

### QCD DGLAP equation: Matrix form

$$
t\,\frac{\partial }{\partial t} \left(
\begin{array}{c}
  q_i(x,t) \\
    g(x,t) \\
\end{array}\right) \,=\,\frac{\alpha_s (t)}{2\,\pi} \sum_{q_j, \bar q_j} \int_x^1\,\frac{d\xi}{\xi}\left(
\begin{array}{cc}
 P_{q_i q_j}\left(\frac{x}{\xi},\alpha_s (t) \right) & P_{q_i g}\left(\frac{x}{\xi},\alpha_s (t) \right) \\
 P_{g   q_j}\left(\frac{x}{\xi},\alpha_s (t) \right) & P_{g g}  \left(\frac{x}{\xi},\alpha_s (t) \right) \\
\end{array}\right)\,\left(
\begin{array}{c}
  q_j(\xi,t) \\
    g(\xi,t) \\
\end{array}\right)
$$

# CNN

Formulation

* **Input:** Height and Width $(H, W)$
* **Filter:** Height and Width $(FH, FW)$
* **Output:** Height and Width $(OH, OW)$
* **Padding:** $P$
* **Stride:** $S$

The height of output images $OH:$

$$ OH \,=\, \frac{H + 2\,P - FH}{S} + 1$$

The width of output images $OW:$

$$ OW \,=\, \frac{W + 2\,P - FW}{S} + 1$$

# Simple plotting example

This is a plot of the $sinc$ function

$$f(x)\,=\,\frac{\sin(x)}{x}$$

```python
%matplotlib inline
%config InlineBackend.figure_format ='svg'

import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)
plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size':13})

plt.yticks([0, 1],['$0$', '$1$'])
plt.xticks([-4*pi, -2*pi, 0, 2*pi, 4*pi],
          ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'])
#plt.axhline(0.5, color='red', lw=0.5)
plt.axhline(0, color='green', lw=0.5)
plt.plot(x, np.sin(x)/x, label=r'$f(x)\,=\,\frac{\sin(x)}{x}$')
plt.ylim(-0.5, 1.1)
plt.title('The $sinc$ function')
plt.legend(loc='best', fontsize=15)
plt.show()
```
