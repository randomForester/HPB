```python
import pandas as pd
import matplotlib.pyplot as plt
#%config InlineBackend.figure_format ='svg'

df = pd.read_csv('UN_CO2_emmisions_data_kilotonne.csv',index_col='Year', parse_dates=['Year'])

df.Monaco = pd.to_numeric(df.Monaco, errors='coerce')
```

```python
df.head()
```

```python
df_totals = df.sum(axis=0).sort_values()
df_totals = df.sum(axis=0).sort_values()[:5]
df_totals = df.sum(axis=0).sort_values()[-9:]

ax = df_totals.plot(kind='bar', figsize=(13,7) ,title='CO$_2$ Emission $_{(1990-2014)}$')
ax.set_ylabel('CO$_2$ (ktonnes)')
ax.get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```
